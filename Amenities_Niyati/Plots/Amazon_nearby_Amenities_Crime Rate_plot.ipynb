```python
# Dependencies
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.misc import imread
```

```python
crime = pd.read_csv('../Results/Crime_Rating.csv')
crime = crime[['City Name','Violent Crime Total', 'Property Crime Total']]
crime['Crime Total'] = round(crime['Violent Crime Total'] + crime['Property Crime Total'])
crime = crime.sort_values('Crime Total')[::-1]
crime = crime.reset_index(drop=True)
crime
```

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]
font_size = plt.rcParams['axes.titlesize']
font_weight = plt.rcParams['axes.titleweight']
print(font_size, font_weight)
 
# Prints: [8.0, 6.0]
print ("Current size:", fig_size)

# Set figure width to 12 and height to 9
fig_size[0] = 20
fig_size[1] = 10
font_size = 25
font_weight = 'bold'
plt.rcParams["figure.figsize"] = fig_size

# plt.style.use(['dark_background'])
plt.figure(facecolor='grey')
plt.axes(facecolor='grey')

plt.title("City Crime Rate (2015)", fontsize=font_size, fontweight=font_weight)
plt.ylabel("Crime Rate (per 100,000 people)", fontsize=20, verticalalignment='bottom', horizontalalignment='center')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

p1 = plt.bar(crime['City Name'], crime['Property Crime Total'],width=.5, color='r',
        alpha=0.6, edgecolor='black', linewidth=2)
p2 = plt.bar(crime['City Name'], crime['Violent Crime Total'],width=.5, color='darkorange',
        alpha=0.8, edgecolor='black', linewidth=2)

plt.grid(linestyle = '--', dashes=(5,6))

plt.legend((p1[0], p2[0]), ('Property Crime', 'Violent Crime'),fontsize='x-large')

plt.savefig('Save_Figs/Crime_rate.png')

plt.show()
```
